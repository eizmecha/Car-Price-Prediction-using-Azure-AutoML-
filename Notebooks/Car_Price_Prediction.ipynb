```python
car.replace("?", np.NAN, inplace=True)
```

```python
car.isnull().sum()
```

```python
#Clearing the price data
car['price'] = pd.to_numeric(car['price'],errors='coerce')
meanGroupbypricemake = car.groupby(by=['make'])['price'].mean().reset_index()

car = car.merge(meanGroupbypricemake,on='make', suffixes=['','_right'])
car['price'] = np.where(car['price'].isnull(), car['price_right'], car['price'])
```

```python
#Remove records where number of doors having null
car = car[car['num-of-doors'].notna()] # Remove 'num-of-doors' values.
car.shape
```

```python
#Remove records where horse power having null
car['horsepower'] = pd.to_numeric(car['horsepower'], errors="coerce")
car.shape
```

```python
car.make.value_counts().nlargest(10).plot(kind='bar', figsize=(15,5), color = 'brown')
plt.title("Number of vehicles by make")
plt.ylabel('Number of vehicles')
plt.xlabel('Make')
```

```python
# Engine size vs price
sns.regplot(x="engine-size", y="price", data=car)
```

Here we can see as the engine-size goes up, the price goes up

This indicates a positive correlation between engine size and price. 

If this is between two independent attributes, then we can remove either one of the attribute from dataset as redundency

But here it is between independent attribute(Engine Size)  and target attrbute(Price).

Therefore Engine size is a good predictor of price .
The correlation between 'engine-size' and 'price' and is nearly 0.87
